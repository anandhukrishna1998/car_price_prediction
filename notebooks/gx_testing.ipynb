```python
import json # Path to your saved expectation suite JSON file
import json
from great_expectations.core.expectation_suite import ExpectationSuite


expectation_suite_path = '/home/sachin/DSP/car_price_prediction/gx/expectations/my_ex.json'

# Load the JSON file
with open(expectation_suite_path, 'r') as f:
    expectation_suite_dict = json.load(f)

# Convert the dictionary to an ExpectationSuite object
expectation_suite = ExpectationSuite(**expectation_suite_dict)
```

```python
import pandas as pd
from great_expectations.dataset.pandas_dataset import PandasDataset


df =pd.read_csv('/home/sachin/DSP/airflow-spike-master/input_data/raw-data/data_chunk_618.csv')
# Assuming `df` is your Pandas DataFrame you want to validate
dataset = PandasDataset(df, expectation_suite=expectation_suite)

# Perform validation
results = dataset.validate()
```
